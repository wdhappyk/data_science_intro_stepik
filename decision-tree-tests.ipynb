```python
import pandas as pd
import numpy as np
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.tree import DecisionTreeClassifier, plot_tree
```

```python
data = pd.DataFrame({
    'is_jumping': [1,1,1,1,1,0,1,1,1],
    'is_meows': [0,1,0,0,1,0,0,1,0],
    'its_barking': [0,0,1,0,0,1,0,0,1],
    'animal': ['kangaroo', 'cat', 'dog', 'kangaroo', 'cat', 'dog', 'kangaroo', 'cat', 'dog']
})
data
```

```python
clf = DecisionTreeClassifier(criterion='entropy')
```

```python
X = data.drop(columns=['animal'])
y = data['animal']
```

```python
clf.fit(X, y)
```

```python
plot_tree(
    clf,
    feature_names=list(X),
    class_names=['cat', 'dog', 'kangaroo'],
    filled=True
)
```
